```python
# chrome_path = "C:\\Users\\kmcre\\chromedriver.exe"
# chrome
base_url = "https://search.naver.com/search.naver?sm=tab_hty.top&where=image&query="

chrome_options = webdriver.ChromeOptions()

chrome_options.add_argument("lang=ko_KR")
chrome_options.add_argument("window-size=1920x1080")
```

```python
def selenium_scroll_option():
    SCROLL_PAUSE_SEC = 3
    
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        time.sleep(SCROLL_PAUSE_SEC)
        
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        if new_height == last_height:
            break
        last_height = new_height
```

```python
# 키워드 검색하기

a=input("검색할 키워드를 입력 : ")
image_name = input("저장할 이미지 이름 : ")

# selenium version 4.0.0 low_version
driver = webdriver.Chrome(chrome_options=chrome_options)


# selenium version 4.3.0 upper
# s = Service(chrome_path)
# driver = webdriver.Chrome(service = s, chrome_options=chrome_options)




print("name before", type(driver), driver)
a = a+" -퍼즐쏘일"# 검색어에서 제외하고싶은 내용 
driver.get(base_url+a)
# browser = driver.find_element_by_name("q")
# browser.send_keys(a)
# browser.send_keys(Keys.RETURN)



# 클래스를 찾고 해당 클래스의 src 리스트를 만들자
print()
print("xpath before", type(driver), driver)

selenium_scroll_option() # 스크롤하여 이미지를 많이 확보


selenium_scroll_option()



'''이미지 src요소를 리스트업해서 이미지 url 저장'''
print()
print("css before", type(driver), driver)
images = driver.find_elements_by_css_selector("._image._listImage") #  클래스 네임에서 공백은 .을 찍어줌

# print("전체 다운로드한 이미지 개수 : {}\n중복된 이미지 개수 : {}\n중복 제거 후 총 이미지 개수 : {}"
#       .format(len(images_url), len(images_url)-len(pd.DataFrame(images_url)[0].unique()),len(pd.DataFrame(images_url)[0].unique())))

count = 1
for image in images:
    try:
        image.click()
        time.sleep(2)
        imgUrl = driver.find_element_by_xpath('//*[@id="main_pack"]/section[3]/div/div[2]/div/div/div[1]/div[1]/div/div/div[1]/div[1]/img').get_attribute('src')
        urlretrieve(imgUrl, image_name+'_'+str(count)+'.jpg')
        print("ok",count)
        
        count+=1
    except:
        print("err",count)
        count += 1
#         pass

    
driver.close()    
```
